Lets now try out the **Random Forest** Library to predict the survival of passengers.

## Random Forest

```r
library(randomForest)
```

```r
train_1$Survived = as.factor(train_1$Survived)
```

```r
for (i in colnames(train_1)){
    
    print(paste(i,sum(is.na(train_1[,i]))))
    
    }
```

```r
mean_age = round(mean(na.omit(train_1$Age)))
```

Replacing the **null values** by **mean**

```r
train_1.imputed<- data.frame(train_1)
```

```r
train_1.imputed$Age[is.na(train_1.imputed$Age)] <- mean(train_1.imputed$Age, na.rm = TRUE)
```

```r
# train_1.imputed <- rfImpute(Survived ~ ., train_1)
```

```r
set.seed(222)
train_1.rf <- randomForest(Survived ~ ., train_1.imputed)
print(train_1.rf)
```

```r
model_rf <- randomForest(Survived ~ ., data = train_1.imputed, importance = TRUE)
```

```r
model_rf
```

```r
str(test_1)
```

```r
sum(is.na(test_1))
```

```r
for (i in colnames(test_1)){
    
    print(paste(i,sum(is.na(test_1[,i]))))
    
    }
```

Here, there are 86 missing values in **Age** and 1 in **Fare** in the test set. Let us inspect the fare first.

```r
print(paste('min fare',min(na.omit(test_1$Fare))))
print(paste('max fare',max(na.omit(test_1$Fare))))
print(paste('mean fare',mean(na.omit(test_1$Fare))))
print(paste('median fare',median(na.omit(test_1$Fare))))
```

```r
test_1[is.na(test_1$Fare),]
```

We will now compare the fares of all the passengers belonging to class 3

```r
class_3<-test_1[ which(test_1$Pclass == 3) , ]
```

```r
class_3$Fare
```

```r
print(paste('min fare',min(na.omit(class_3$Fare))))
print(paste('max fare',max(na.omit(class_3$Fare))))
print(paste('mean fare',mean(na.omit(class_3$Fare))))
print(paste('median fare',median(na.omit(class_3$Fare))))
```

Let us replace the missing fare value by the median fare

```r
test_1[is.na(test_1$Fare),]$Fare<- median(na.omit(class_3$Fare))
```

Now it is time to tackle the **missing Age values**. For that purpose we are going to replace the missing values with the mean of all the other Age values, since we made the same assumption for the Training set.

```r
test_1$Age[is.na(test_1$Age)] <- mean(test_1$Age, na.rm = TRUE)
```
